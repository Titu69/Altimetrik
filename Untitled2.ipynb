```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
healthexp = sns.load_dataset('healthexp')
healthexp.head()
```

```python
healthexp.describe()
```

```python
healthexp.info()
```

```python
healthexp.isnull().sum()
```

```python
Year = healthexp['Year']
Country = healthexp['Country']
Spending_USD = healthexp['Spending_USD']
Life_Expectancy = healthexp['Life_Expectancy']
```

```python
sns.scatterplot(x='Life_Expectancy', y='Spending_USD', data=healthexp, hue='Year')
plt.title('Spending_USD vs Life_Expectancy')
plt.show()
```

```python
sns.lineplot(x='Year', y='Life_Expectancy', data=healthexp, hue='Country')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Life Expectancy (Years)", fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x="Country", y="Life_Expectancy", data=healthexp)
plt.title("Average Life Expectancy by Country", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Life_Expectancy", fontsize=12)
plt.show()

plt.figure(figsize=(5, 3))
sns.barplot(x="Country", y="Spending_USD", data=healthexp)
plt.title("Average Spending by Country", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Spending_USD", fontsize=12)
plt.show()
```

```python
correlation_data = healthexp[["Year", "Spending_USD", "Life_Expectancy"]]
```

```python
# Compute correlation matrix
corr_matrix = correlation_data.corr(method='pearson')  # You can also use "spearman"
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

# Add title
plt.title("Correlation Matrix: Year, Spending, Life Expectancy", fontsize=14)
plt.tight_layout()
plt.show()
```
